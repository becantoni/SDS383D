```python
#A)
#%%simulate data
X = np.linspace(0,50, 1000)
y = []
for i in range(len(X)):
    y.append(np.random.normal(loc= np.sin(0.2*X[i]), scale = 0.1))
#subtract mean
y = y - np.mean(y)
X = X - np.mean(X) 
plt.scatter(X, y)
```

```python
#perform comparison based on mse
from kernel_regression import kernel_regression
reg = kernel_regression(X,y)
reg.h_train()  
```

```python
#B)
#simulate new dataset and plot it
X_cell = np.linspace(0,20, 500)
y_cell=[]
#%%
#wiggly function, small noise
cell_y = []
for i in range(len(X_cell)):
    cell_y.append(np.random.normal(loc= np.sin(2*X_cell[i]), scale =.1))
plt.scatter(X_cell, cell_y)  
plt.show()
y_cell.append(np.asarray(cell_y))

#wiggle function, high noise
cell_y = []
for i in range(len(X_cell)):
    cell_y.append(np.random.normal(loc= np.sin(2*X_cell[i]), scale = np.sqrt(0.5)))
plt.scatter(X_cell, cell_y)
y_cell.append(np.asarray(cell_y))
plt.show()

#smooth function, small noise
cell_y = []
for i in range(len(X_cell)):
    cell_y.append(np.random.normal(loc= np.sin(0.2*X_cell[i]), scale = .1))
y_cell.append(np.asarray(cell_y))
plt.scatter(X_cell, cell_y)  
plt.show()

#smooth function, high noise
cell_y = []
for i in range(len(X_cell)):
    cell_y.append(np.random.normal(loc= np.sin(0.2*X_cell[i]), scale = np.sqrt(0.5)))
plt.scatter(X_cell, cell_y)
y_cell.append(np.asarray(cell_y))
plt.show()
```

```python
#perform cross validation on previous dataset
#%% apply method to the four datasets
for t in range(len(y_cell)):
    fit = kernel_regression(X_cell, y_cell[t])
    fit.h_train() 
```
